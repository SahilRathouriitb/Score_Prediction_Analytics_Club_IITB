# SOLVING THE PROBLEM OF MATCH ID AND INNINGS

```python
# The problem which we have been facing is that the same match id is representing both the innings 
# since we have to process both the innings seperately most of the times , so it is causing problems 
# we shall combine mactch id and innings to make a unique match id 
```

```python
df_f.head()
```

```python
# to give unique match id to both innings of the match we can add .1 or .2 as decimal for 1st and 2nd innings respectively
df_f['innings']=df_f['innings']*100000000
```

```python
df_f['ID']=(df_f['match_id']+df_f['innings'])
```

```python
df_f.drop(columns=['match_id'],inplace=True)
```

# Runs in last 5 over

```python
df_f.shape
```

```python

match_ids = df_f['ID'].unique()
```

```python
match_ids
```

```python
last_five = []

for match_id in match_ids:
    # Filter the DataFrame for the current match_id
    match_data = df_f[df_f['ID'] == match_id]

    # Use rolling window to calculate the cumulative sum over a window of 30 balls
    cumulative_sum = match_data['ball_runs'].rolling(window=30).sum()

    # Extend the last_five list with the cumulative sum values
    last_five.extend(cumulative_sum.values.tolist())
```

```python
df_f['last_five'] = last_five
```

```python
df_f.head(30)
```

```python
# also we have to construct a column called ['Total Score']
final_df = df_f.groupby('ID').sum()['ball_runs'].reset_index().merge(df_f,on='ID')
```

# FEATURE TRANSFORMATION

```python
# we will use pipelines and columns transformers to perform this task
'''1-Onehotencoding 
2-Scaling using MinMaxScaling
3-Function Transformers or Power Transformers 
3-Outlier Detection'''
```

```python
X_train.head()
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,MinMaxScaler,PowerTransformer
from sklearn.ensemble import RandomForestRegressor
import numpy as np
from sklearn.metrics import r2_score
# we will make seperate transformers and connect them with pipelines 
```

```python
l=[]
for i in range(0,24):
    l.append(i)
    # i will be using this list in one of the column transformer 
```

```python
X_train
```

```python
trf1=ColumnTransformer(transformers=[('encoding',OneHotEncoder(drop='first',sparse_output=False),[1,2])],remainder='passthrough')
trf2=ColumnTransformer(transformers=[('scaling',MinMaxScaler(),[18,19,20,21,22,23])],remainder='passthrough')
trf3=ColumnTransformer(transformers=[('transforming',PowerTransformer(method='yeo-johnson'),l)],remainder='passthrough')
```

```python
# ONE MORE THING I NOTICED THAT MY DATA IS HIGHLY RIGHT SKEWED SO I NEED TO APPLY EITHER LOG TRNASFORM OR YEO JOHNSON TRANSFORM
# since i have a lot of 0's in my data box-cox is ruled out also i will not use simple log , i will go for log1p
```

```python
mp=make_pipeline(trf1,trf2,trf3)
```

```python
X_train_f=mp.fit_transform(X_train)
X_test_f=mp.transform(X_test)
#Y_pred=mp.predict(X_test)
```

# K-FOLD STACKING STRUCTURE (K=5 Default)

```python
from sklearn.ensemble import StackingRegressor,RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score
```

```python
base_model=[('knn',KNeighborsRegressor(n_neighbors=5,weights='distance',n_jobs=-1)),
            ('rf',RandomForestRegressor(bootstrap=False,
 max_features=0.25,
 max_samples=None,
 min_samples_leaf=1,
 min_samples_split=2,
 n_estimators=150))]
```

```python
sr=StackingRegressor(estimators=base_model,final_estimator=XGBRegressor(n_estimators=200,learning_rate=0.1))
```

```python
sr.fit(X_train_f,Y_train)
Y_pred=sr.predict(X_test_f)
print(r2_score(Y_pred,Y_test))
```
